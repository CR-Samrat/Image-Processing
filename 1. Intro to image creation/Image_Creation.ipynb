```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

## Creation of a small black & white image (4 * 4)

```python
image_1 = np.array([[0,0,0,0],
                   [255,255,255,255],
                   [0,0,0,0],
                   [255,255,255,255]])

# 0   means black
# 255 means white
```

```python
plt.imshow(image_1)
cv.imwrite("small_bw.png",image_1)
```

## Creation of a large black & white image (256 * 256)

```python
image_2 = np.zeros((256,256))

image_2
```

```python
for i in range(image_2.shape[0]):
    for j in range(image_2.shape[1]):
        image_2[i][j] = i

image_2
```

```python
plt.imshow(image_2)
cv.imwrite("large_bw.png",image_2)
```

## Creation of a simple black & white image in RGB pallet (4 * 4)

```python
mapping = {
    0 : [0,0,0],
    255 : [255, 255, 255]
}

image_3 =[[0,255,0,255],
        [255,0,255,0],
        [0,255,0,255],
        [255,0,255,0]]

for i in range(len(image_3)):
    for j in range(len(image_3[0])):
        image_3[i][j] = mapping[image_3[i][j]]

image_3
```

```python
plt.imshow(image_3)
cv.imwrite("small_rgb.png",np.array(image_3))
```
