```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

## Loading an image using OpenCV

```python
image = cv.imread('rgb.jpeg')

plt.imshow(image)
```

## Converting BGR image into RGB

```python
image = cv.cvtColor(image, cv.COLOR_BGR2RGB)

plt.imshow(image)
```

## Image Slicing

```python
image.shape
```

```python
image_1 = image[:, :, :]

plt.imshow(image_1)
```

Slicing by height

```python
image_2 = image[:100, :, :]

plt.imshow(image_2)
```

```python
image_3 = image[100:200, :, :]

plt.imshow(image_3)
```

Slicing by width

```python
image_4 = image[:, 100:300, :]

plt.imshow(image_4)
```

Slicing by both side

```python
image_5 = image[30:190, 100:300, :]

plt.imshow(image_5)
```

## Frame Extraction

```python
# extracting red
red_freg = image[:,:,0]

# extracting green
green_freg = image[:,:,1]

# extracting blue
blue_freg = image[:,:,2]
```

```python
plt.imshow(red_freg)
cv.imwrite('red_fragment.jpg',red_freg)
```

```python
plt.imshow(green_freg)
cv.imwrite('green_fragment.jpg',green_freg)
```

```python
plt.imshow(blue_freg)
cv.imwrite('blue_fragment.jpg',blue_freg)
```

## Displaying image using OpenCV

```python
cv.imshow("Real image",image)
cv.waitKey(0)
cv.imshow("Red fragment",red_freg)
cv.waitKey(0)
cv.imshow("Green fragment",green_freg)
cv.waitKey(0)
cv.imshow("Blue image",blue_freg)
cv.waitKey(0)
```
