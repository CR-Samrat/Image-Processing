```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

## Creating different shades of red

```python
image_1 = np.array([
    [[30,0,0],[60,0,0],[90,0,0]],
    [[120,0,0],[150,0,0],[180,0,0]],
    [[210,0,0],[230,0,0],[255,0,0]]
])

plt.imshow(image_1)
cv.imwrite("red.png",image_1)
```

## Creating different shades of blue

```python
image_3 = np.array([
    [[0,0,30],[0,0,60],[0,0,90]],
    [[0,0,120],[0,0,150],[0,0,180]],
    [[0,0,210],[0,0,230],[0,0,255]]
])

plt.imshow(image_3)
cv.imwrite("blue.png",image_3)
```

## Creating dense red (256 * 256)

```python
image_4 = []

for i in range(256):
    temp = []
    for j in range(256):
        temp.append([i,0,0])
    image_4.append(temp)

image_4 = np.array(image_4)
```

```python
plt.imshow(image_4)
cv.imwrite("dense_red.png",image_4)
```

## Mixup dense image

```python
image_5 = []

for i in range(256):
    temp = []
    for j in range(1536):
        temp.append([i,0,0])
    image_5.append(temp)
for i in range(255,0,-1):
    temp = []
    for j in range(1536):
        temp.append([i,0,0])
    image_5.append(temp)
for i in range(256):
    temp = []
    for j in range(1536):
        temp.append([0,i,0])
    image_5.append(temp)
for i in range(255,0,-1):
    temp = []
    for j in range(1536):
        temp.append([0,i,0])
    image_5.append(temp)
for i in range(256):
    temp = []
    for j in range(1536):
        temp.append([0,0,i])
    image_5.append(temp)
for i in range(255,0,-1):
    temp = []
    for j in range(1536):
        temp.append([0,0,i])
    image_5.append(temp)

image_5 = np.array(image_5)
```

```python
plt.imshow(image_5)
cv.imwrite("dense_rgb.png",image_5)
```
